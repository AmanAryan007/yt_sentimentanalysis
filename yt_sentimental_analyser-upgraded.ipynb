```python
%%writefile my_app.py

import streamlit as st
from googleapiclient.discovery import build
from textblob import TextBlob

# Function to retrieve video comments
def get_video_comments(service, **kwargs):
    comments = []
    results = service.commentThreads().list(**kwargs).execute()

    while results:
        for item in results['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comments.append(comment)

        if 'nextPageToken' in results:
            kwargs['pageToken'] = results['nextPageToken']
            results = service.commentThreads().list(**kwargs).execute()
        else:
            break

    return comments

# Function to analyze sentiment and classify as positive, negative, or neutral
def analyze_sentiment(text):
    analysis = TextBlob(text)
    sentiment = analysis.sentiment.polarity
    if sentiment > 0:
        return "Positive"
    elif sentiment < 0:
        return "Negative"
    else:
        return "Neutral"

# Set up the YouTube API key and build the API client
api_key = st.text_input('Enter your YouTube API Key', 'YOUR_API_KEY')
youtube = build('youtube', 'v3', developerKey=api_key)

# Streamlit interface for YouTube comments and article analysis
st.title('Sentiment Analysis Application')

# YouTube Comments Analysis
st.header('YouTube Comment Sentiment Analysis')

video_id = st.text_input('Enter YouTube Video ID', 'NqTysMyLuTA')

if st.button('Analyze YouTube Comments'):
    # Call the function to retrieve the comments
    video_comments = get_video_comments(youtube, part='snippet', videoId=video_id, textFormat='plainText')

    total_sentiment_score = 0
    comment_count = len(video_comments)

    st.write(f"Number of Comments: {comment_count}")

    for idx, comment in enumerate(video_comments, start=1):
        sentiment = analyze_sentiment(comment)
        st.write(f"Comment {idx}: {comment} (Sentiment: {sentiment})")
        total_sentiment_score += TextBlob(comment).sentiment.polarity

    # Calculate average sentiment score
    average_sentiment_score = total_sentiment_score / comment_count if comment_count > 0 else 0

    st.write(f"Total Sentiment Score: {total_sentiment_score}")
    st.write(f"Average Sentiment Score: {average_sentiment_score}")

# Article Sentiment Analysis
st.header('Article Sentiment Analysis')

article = st.text_area('Enter the article text here')

if st.button('Analyze Article Sentiment'):
    article_sentiment = analyze_sentiment(article)
    st.write(f"Article Sentiment: {article_sentiment}")
```
